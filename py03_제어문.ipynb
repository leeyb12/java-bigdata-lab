```python
# 로또 번호
# 1 ~ 45임의의 수 중에서 중복 없는 6개의 수를 뽑아 나열하여 분석하시오.

import random as ran 
lotto = [] # 로또 번호 저장할 리스트 
n = 0 # 뽑은 횟수 카운트 

# 중복 없는 6개 번호 뽑기 
while n < 6: 
    num = ran.randint(1, 45) 
    if num not in lotto: # 중복 체크 
        lotto.append(num) 
        n += 1 
        
print("이번 로또 번호:", lotto)
```

```python
import random as ran

# 출현 횟수를 저장할 리스트 (1~45)
count = [0] * 46  # 인덱스 0은 사용하지 않음

draws = 2000 # 반복 횟수
i = 0         # 전체 추첨 횟수 카운트

while i < draws:
    lotto = []  # 로또 번호 저장할 리스트
    n = 0       # 뽑은 횟수 카운트

    # 중복 없는 6개 번호 뽑기
    while n < 6:
        num = ran.randint(1, 45)
        if num not in lotto:  # 중복 체크
            lotto.append(num)
            n += 1
            count[num] += 1   # 출현 횟수 기록
    i += 1

# 가장 많이 출현한 숫자 찾기
max_count = max(count)
most_common_nums = []

j = 1
while j <= 45:
    if count[j] == max_count:
        most_common_nums.append(j)
    j += 1
print(lotto)
print(f"총 {draws}회 추첨 결과")
print("가장 많이 출현한 숫자:", most_common_nums)
print(f"출현 횟수: {max_count}회")
```

```python
# 주사위의 눈금 조작. 3번을 16.x -> 25%로 확률을 조작하라.
import random as ran
import matplotlib.pyplot as plt
box = []
n = 0
cc = 10000
choice = int(input("조작할 확률 %를 입력하세요.") or 10) # 퍼센트 조절
my = int(input("조작할 번호를 입력하세요.") or 3)
values = []
while n < cc:
    num = ran.randint(1, 6)
    lucky = ran.randint(1, 100)
    if choice > lucky:
        box.append(my)
    else:
        box.append(num)
    n += 1
if cc == len(box):
    print(f"{cc}회 정상 진행 되었습니다.")
    for n in range(1, 7):
        print(f"주사위 눈금 {n}이 {box.count(n)}번 { ((box.count(n)/cc) * 100):0.2f}% 출현하였습니다.")
        values.append(box.count(n))
else:
    print(f"검증 결과가 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")
color = ["skyblue", "orange", "red", "purple", "green", "aqua"]
plt.bar(range(1, 7),values, color=color)
plt.show()
```
